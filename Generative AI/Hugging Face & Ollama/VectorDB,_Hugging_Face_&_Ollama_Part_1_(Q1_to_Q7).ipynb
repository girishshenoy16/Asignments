 1) What is a Vector Database (VectorDB) and how is it different from traditional databases?

->

 🔹 What is a Vector Database?

A Vector Database (VectorDB) is a specialized database designed to store, index, and retrieve vector embeddings efficiently.

Vector embeddings are high-dimensional numerical representations of data such as:

- Text (sentences, documents)
- Images
- Audio
- Video
- User behavior

These embeddings are generated using deep learning models like BERT, Sentence Transformers, or CLIP.

Example:

When a sentence like:

> "Machine learning is transforming industries."

is passed through an embedding model, it becomes a vector like:


[0.12, -0.98, 0.45, ..., 0.33]


Each number represents semantic information captured by the model.


 🔹 Why Do We Need Vector Databases?

Traditional databases are optimized for:

- Exact matches
- Structured data
- Relational queries

However, AI applications require:

- Similarity search
- Semantic matching
- Nearest neighbor search in high-dimensional space

Vector databases solve this problem.


 🔹 How VectorDB Works

1. Convert data into embeddings using ML models.
2. Store embeddings in the vector database.
3. When a query comes:

   - Convert query into embedding.
   - Perform similarity search.
   - Retrieve closest matching vectors.

Similarity metrics include:

- Cosine Similarity
- Euclidean Distance
- Dot Product



 🔹 Difference Between Vector Databases and Traditional Databases

| Feature         | Traditional Database       | Vector Database                |
| --------------- | -------------------------- | ------------------------------ |
| Data Type       | Structured (rows, columns) | High-dimensional vectors       |
| Query Type      | Exact matching (SQL)       | Similarity search              |
| Indexing        | B-tree, Hash index         | ANN, HNSW, IVF                 |
| Example Systems | MySQL, PostgreSQL          | Pinecone, Chroma, FAISS        |
| Use Cases       | Banking, ERP, CRM          | RAG, AI search, recommendation |



 🔎 Key Concept Difference

Traditional Query:

sql
SELECT - FROM users WHERE id = 101;


Vector Query:

> "Find documents most similar to this paragraph."



 🔹 Real-World Applications

- Chatbots (Retrieval-Augmented Generation)
- Recommendation systems
- Fraud detection
- Image similarity search
- Voice assistants

Vector databases are a core infrastructure component in modern AI systems.

```python
"""
6) Install and set up Chroma DB, and insert sample vector data for semantic
search.

->

"""

import chromadb
from chromadb.config import Settings

# Initialize the Persistent Client
# This saves your data to a local folder named 'chroma_storage'
client = chromadb.PersistentClient(path="./chroma_storage")

# Create or Get a Collection
# A collection is like a table in a database
collection = client.get_or_create_collection(name="my_knowledge_base")

# Add Documents with Metadata and IDs
# Chroma automatically generates embeddings if you don't provide them
print("Inserting sample data...")
collection.add(
    documents=[
        "Python is a versatile programming language for AI development.",
        "The Eiffel Tower is located in Paris, France.",
        "Vector databases like Chroma are essential for RAG systems.",
        "A croissant is a buttery, flaky pastry from France.",
        "Machine learning models require clean datasets for training."
    ],
    metadatas=[
        {"source": "tech_blog"},
        {"source": "travel_guide"},
        {"source": "ai_docs"},
        {"source": "food_wiki"},
        {"source": "ml_handbook"}
    ],
    ids=["doc1", "doc2", "doc3", "doc4", "doc5"]
)

# Perform a Semantic Search
# Notice we don't use the word 'Python' or 'Language', but Chroma finds the match
query = "Tell me about coding for artificial intelligence"
print(f"\nSearching for: '{query}'")

results = collection.query(
    query_texts=[query],
    n_results=2  # Number of top matches to return
)

# Display Results
for i in range(len(results['ids'][0])):
    print(f"\nMatch {i+1}:")
    print(f"ID: {results['ids'][0][i]}")
    print(f"Document: {results['documents'][0][i]}")
    print(f"Metadata: {results['metadatas'][0][i]}")
    print(f"Distance (lower is better): {results['distances'][0][i]}")
```

```python
"""
7)  Demonstrate how to download and fine-tune a Hugging Face model for a
text generation task.

->

"""
from transformers import AutoTokenizer, AutoModelForCausalLM, Trainer, TrainingArguments, pipeline
from datasets import load_dataset

model_name = "gpt2"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Fix padding issue in GPT-2
tokenizer.pad_token = tokenizer.eos_token
model.config.pad_token_id = tokenizer.eos_token_id

dataset = load_dataset("wikitext", "wikitext-2-raw-v1")

# Remove empty rows
dataset = dataset.filter(lambda x: len(x["text"]) > 0)

def tokenize(example):
    tokens = tokenizer(
        example["text"],
        truncation=True,
        padding="max_length",
        max_length=128
    )

    tokens["labels"] = tokens["input_ids"].copy()
    return tokens

tokenized = dataset.map(tokenize)

train_dataset = tokenized["train"].select(range(2000))


training_args = TrainingArguments(
    output_dir="./results",
    per_device_train_batch_size=2,
    num_train_epochs=1,
    logging_steps=50
)


trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset
)

trainer.train()
```

```python
generator = pipeline("text-generation", model=model, tokenizer=tokenizer)

result = generator(
    "Artificial Intelligence will",
    max_length=50,
    num_return_sequences=1
)

print(result)
```
